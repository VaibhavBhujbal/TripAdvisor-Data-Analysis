```python
import requests
import bs4
import pandas as pd
locations=[]
names=[]
times=[]
reviews=[]
subjects=[]
i=0
#Scrapping is kept limited for 2000
while i<2000:
    resp= requests.get("https://www.tripadvisor.com.sg/Hotel_Review-or"+str(i)+"-g294265-d1770798-Reviews-Marina_Bay_Sands-Singapore.html")
    print("https://www.tripadvisor.com.sg/Hotel_Review-or"+str(i)+"-g294265-d1770798-Reviews-Marina_Bay_Sands-Singapore.html")
    lxml_soup=bs4.BeautifulSoup(resp.text,'lxml')
    userinfo=lxml_soup.select(".info_text")

    if(lxml_soup.select(".noQuotes")):
        subj=lxml_soup.select(".noQuotes")
        
        for s in subj:
            subjects.append(s.text)
    else:
        subjects.append("None")


    for info in userinfo:
        if (info.div):
            names.append((info.div).text)
        else:
            names.append("None")
        if (info.select(".userLoc")):
            infs=info.select(".userLoc")
            for inf in infs:
                locations.append(inf.text)
        else:
            locations.append("None")
        
    timeinfo=lxml_soup.select(".ratingDate")
    for t in timeinfo:
        times.append(t.text)

    if(lxml_soup.select(".partial_entry")):
        rev=lxml_soup.select(".partial_entry")
        for r in rev:
            reviews.append(r.text)
    else:
        reviews.append("None")
    
    i=i+5
            
```

```python
#This will converge all lists into dictionary and write to CSV
data_dict = {'subject': subjects, "name": names, "review": reviews, "time": times, "location": locations}
df = pd.DataFrame.from_dict(data_dict, orient='index')
df = df.transpose()
df.to_csv("test.csv")
df
```
